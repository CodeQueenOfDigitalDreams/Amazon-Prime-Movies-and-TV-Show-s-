```python
df=pd.read_csv('/kaggle/input/amazon-prime-movies-and-tv-shows/amazon_prime_titles.csv')

#lets check head of data to see how data has been defined
df.head()
```

```python
#lets check tail of data to see how data has been defined
df.tail()
```

```python
df['show_id'] = df['show_id'].str.replace('s', '')
df.head()
```

```python
#lets check name of all colums
c=df.columns
print(c)
```

```python
#now lets check how many NAN values are in all columns
q=df.isnull().sum()*100/len(df)
print(q)

#now we can see % director,cast,country,date_added & rating has majorly NAN vlaue
```

```python
#now we are deleting 2 columns (date_added, country)
df.drop(columns=['country', 'date_added'],inplace=True)
```

```python
df.head()
```

```python
#now we will see columns by column how we can fix these NAN values
#start with director as it has only 21%
#for that we need to see its unique values
pd.set_option('display.max_columns',None)
w=df['director'].isna().sum()
print(w)
```

```python
#Now we can see it has more NAN value but not that much for which we will delete the whole colum 
#and it does not have that less % that we can remove that particular rows
#therefore we left with the option to fill those columns as Unknown
e=df.fillna({'director':'Unknown'},inplace=True)
print(e)
#Now in director columns we can see there is no more NAN values itstead there is unknown
```

```python
#now we will see next column how we can fix these NAN 
#start with Cast as it has only 12%
#for that we need to see its unique values
pd.set_option('display.max_columns',None)
w=df['cast'].isna().sum()
print(w)
#we will do the same with it as director
```

```python
#Now we can see it has more NAN value but not that much for which we will delete the whole colum 
#and it does not have that less % that we can remove that particular rows
#therefore we left with the option to fill those columns as Unknown
#usually we can do it together but in this for your understanding we did it in different code 
e=df.fillna({'cast':'Unknown'},inplace=True)
print(e)
#Now in director columns we can see there is no more NAN values itstead there is unknown
```

```python
#now last but not the least
#we have to fix rating columns
#as in rating colum have very less number of % of NAN we will only delete those rows
r=df.dropna(subset=['rating'],inplace=True)
print(r)
```

```python
k=df.isnull().sum()*100/len(df)
print(k)
#now as we can see we do not have any NAN values in any columns
```

```python
#now lets see rating uniques values
t=df['rating'].unique()
print(t)
```

```python
#as we have seen some values are same but with different name so we can make them same
#basically we are replacing the names to make it less cluster
df['rating']=df['rating'].replace({
    "16": "16+",
    "AGES_16_": "16+",
    "AGES_18_": "18+",
    "R": "18+",
    "NC-17": "18+",
    "13+": "PG-13",
    "PG-13": "PG-13",
    "G": "GENERAL",
    "ALL": "GENERAL",
    "ALL_AGES": "GENERAL",
    "UNRATED": "UNRATED",
    "NOT_RATE": "UNRATED",
    "NR": "UNRATED",
    "7+": "TV-Y7",
    "TV-NR": "TV-UNRATED"    
})
```

```python
u=df['duration'].unique()
print(u)
```

```python
df['duration_by_min']=df['duration'].str.split('min',expand=True)[0]
df.head()
```

```python
#now we have define a condition and we will apply it in pandas
#in condition we said if in x there is term 'min' then split it otherwise give 0
def condi(x):
    if 'min' in x:
        return int(x.split()[0])
    else:
        return 0
df['duration_in_min']=df['duration'].apply(condi)

df.head()
```

```python
#we are doing same with season
#now we have define a condition and we will apply it in pandas
#in condition we said if in x there is term 'season' then split it otherwise give 0
def condi(x):
    if 'season' in x:
        return int(x.split()[0])
    else:
        return 0
df['duration_in_seasons']=df['duration'].apply(condi)

df.head()
```

```python
#as we have already splitted duration into min and season we do not need that colum so we will delete it
u=df.drop(columns=['duration'],inplace=True)
df.head()
```

```python
#as with description of movie we can not create any visualisation so we will drop that too 
#however it is not nessary
df.drop(columns=['description'],inplace=True)
df.head()
```

```python
col=df.columns
print(col)
a=df.astype({'show_id':'int', 'release_year':'int',})
```

```python
#while saving csv file we hav added index because our show_id has some unevens becasue we have deleted some rows
df.to_csv("/kaggle/working/amazon_prime.csv",index=True,encoding="utf-8")
```
